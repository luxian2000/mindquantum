```python
import mindspore as ms
import numpy as np
from mindspore import nn, ops, Tensor, Parameter
```

```python
np.random.seed(42)
x = ops.ones(5, ms.float32) + 1
y = ops.zeros(3, ms.float32)
w = Parameter(Tensor(np.random.randn(5,3), ms.float32), name='w')
b = Parameter(Tensor(np.random.randn(3,), ms.float32), name='b')
```

```python
def function(x, y, w, b):
    z = ops.matmul(x, w) + b
    return z
```

```python
print(function(x, y, w, b))
```

```python
g = ms.grad(function, (2,3))
```

```python
g(x, y, w, b)
```

```python
w -= g(x, y, w, b)[0]
```
